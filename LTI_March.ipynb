```python
### Build agents
from langchain.agents import initialize_agent, AgentType
from langchain.tools import Tool
from langchain_openai import ChatOpenAI
```

```python
from google.colab import userdata
import os
os.environ["OPENAI_API_KEY"]=userdata.get('openai')
```

```python
llm=ChatOpenAI()
```

```python
def calculator_tool(input_text: str) -> str:
    try:
        return str(eval(input_text))
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
tools=[Tool(name="Calculator",func=calculator_tool,description="Useful for solving math problems")]
```

```python
agent=initialize_agent(tools=tools,llm=llm,agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,verbose=True)
```

```python
response = agent.run("What is (12 * 5) + 20?")
print(response)
```

```python
from langchain.memory import ConversationBufferMemory
memory = ConversationBufferMemory(memory_key="chat_history")
agent=initialize_agent(tools=tools,llm=llm,agent=AgentType.CONVERSATIONAL_REACT_DESCRIPTION,memory=memory,verbose=True)
```

```python
agent.run("My name is Premalatha")
```

```python
agent.run("What is 50+70")
```

```python
agent.run("What is my name?")
```

```python
### Multiagents
```

```python
def calculator_tool(input_text: str) -> str:
    try:
        return str(eval(input_text))
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
def word_count_tool(text: str) -> str:
    return f"Word count: {len(text.split())}"
```

```python
multi_tools=[Tool(name="MAthAgent",func=calculator_tool,description="Useful for math"),
             Tool(name="TextAgent",func=word_count_tool,description="Count no of words")]
```

```python
supervisor_agent=initialize_agent(tools=multi_tools,llm=llm,agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,verbose=True)
```

```python
supervisor_agent("What is 25 *4 +10")
```

```python
supervisor_agent("Count words in 'AI is transforming industries' ")
```
